#**Image to Pencil Sketch with Python**

## **Grayscale Image:**

For a grayscale image, BackgroundTreshold is set to (0,0,0) by default  

## **Inverted Image:**

The image focussed directly on the retina of the eye is an inverted image.

## **Blurred image:**

If a blurred image is observed carefully then a common thing to notice is that image is smooth meaning edges are not observed.

```python
import cv2
import warnings
import matplotlib.pyplot as plt
import numpy as np


warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
#read image
image = cv2.imread("/content/dog_Image.png")
#set title of Image
plt.title("Dog Image")
plt.imshow(image)
plt.show()
```

# convert to gray image

```python
# convert to gray image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.title("Dog Gray Image")
plt.imshow(gray_image)
plt.show()
```

# Get Inverted Image

```python
inverted_image = 255- gray_image
plt.title("Inverted Image")
plt.imshow(inverted_image)
plt.show()
```

#convert image to Blur

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
inverted_blurred = 255 - blurred# invert the blur image
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)#convert to pencil sketch
plt.title("Pencil sketch Image")# show title
plt.imshow(pencil_sketch)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(image)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
